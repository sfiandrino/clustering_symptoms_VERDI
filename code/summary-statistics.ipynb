```python
def get_statistics(df, dict_stat, stat_cols):
    for col in stat_cols:
        if (col == 'num_symptoms') or (col == 'median_sym_duration') or (col == 'num_comorbidities') :
            median = df[col].median()
            irq25 = round(df[col].quantile(0.25),3)
            irq75 = round(df[col].quantile(0.75),3)
            irq = str(irq25) + '-' + str(irq75)
            colname = col + ' - Median (IRQ 25-75)' 
            dict_stat[colname] = str(round(median, 3)) + ' (' + str((irq)) + ') '
            if col == 'num_comorbidities':
                #Comorbities 0
                dfRC_com0 = df[df[col] == 0]
                num = dfRC_com0.shape[0]
                perc = dfRC_com0.shape[0] / df.shape[0] * 100
                colname_com = col + '0 - Num (Perc)'
                dict_stat[colname_com] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
                
                #Comorbidities 1
                dfRC_com1 = df[df[col] == 1]
                num = dfRC_com1.shape[0]
                perc = dfRC_com1.shape[0] / df.shape[0] * 100
                colname_com1 = col + '1 - Num (Perc)'
                dict_stat[colname_com1] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
                
                #Comorbidities more than 1
                dfRC_com2 = df[df[col] > 1]
                num = dfRC_com2.shape[0]
                perc = dfRC_com2.shape[0] / df.shape[0] * 100
                colname_com2 = col + 'More than 1 - Num (Perc)'
                dict_stat[colname_com2] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
            elif col == 'num_symptoms':
                #Symptoms 0
                dfRC_sym0 = df[df[col] == 0]
                num = dfRC_sym0.shape[0]
                perc = dfRC_sym0.shape[0] / df.shape[0] * 100
                colname_sym = col + '0 - Num (Perc)'
                dict_stat[colname_sym] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
                
                #Symptoms 1
                dfRC_sym1 = df[df[col] == 1]
                num = dfRC_sym1.shape[0]
                perc = dfRC_sym1.shape[0] / df.shape[0] * 100
                colname_sym1 = col + '1 - Num (Perc)'
                dict_stat[colname_sym1] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
                
                #Symptoms 2
                dfRC_sym2 = df[df[col] == 2]
                num = dfRC_sym2.shape[0]
                perc = dfRC_sym2.shape[0] / df.shape[0] * 100
                colname_sym2 = col + '2 - Num (Perc)'
                dict_stat[colname_sym2] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
                
                #Symptoms More than 3
                dfRC_sym3 = df[df[col] > 2]
                num = dfRC_sym3.shape[0]
                perc = dfRC_sym3.shape[0] / df.shape[0] * 100
                colname_sym3 = col + '>= 3 - Num (Perc)'
                dict_stat[colname_sym3] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
        elif col == 'age_enrollment': 
            median = df[col].median()
            irq25 = round(df[col].quantile(0.25),3)
            irq75 = round(df[col].quantile(0.75),3)
            irq = str(irq25) + '-' + str(irq75)
            colname = col + ' - Median (IRQ 25-75)'
            dict_stat[colname] = str(round(median, 3)) + ' (' + str(irq) + ') '
            
            #0-2 year
            df_RC_02y = df[df[col] <= 2]
            num = df_RC_02y.shape[0]
            perc = df_RC_02y.shape[0] / df.shape[0] * 100
            colname02 = col + '0-2 year - Num (Perc)'
            dict_stat[colname02] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
            
            #3-5 year
            df_RC_35y = df[(df[col] >= 3) & (df[col] <=5)]
            num = df_RC_35y.shape[0]
            perc = df_RC_35y.shape[0] / df.shape[0] * 100
            colname35 = col + '3-5 year - Num (Perc)'
            dict_stat[colname35] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
            
            #6-10 year
            df_RC_610y = df[(df[col] >= 6) & (df[col] <=10)]
            num = df_RC_610y.shape[0]
            perc = df_RC_610y.shape[0] / df.shape[0] * 100
            colname610 = col + '6-10 year - Num (Perc)'
            dict_stat[colname610] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
            
            #11-13 year
            df_RC_1113y = df[(df[col] >= 11) & (df[col] <=13)]
            num = df_RC_1113y.shape[0]
            perc = df_RC_1113y.shape[0] / df.shape[0] * 100
            colname1113 = col + '11-13 year - Num (Perc)'
            dict_stat[colname1113] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
            
            #14-20 year
            df_RC_1420y = df[(df[col] >= 14) & (df[col] <=20)]
            num = df_RC_1420y.shape[0]
            perc = df_RC_1420y.shape[0] / df.shape[0] * 100
            colname1420 = col + '14-20 year - Num (Perc)'
            dict_stat[colname1420] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
            
        else:
            list_unique = df[col].unique().tolist()
            for j in range(len(list_unique)):
                num = (df[col] == list_unique[j]).sum()
                perc = (df[col] == list_unique[j]).sum() / df[col].shape[0] * 100
                colname = col + ' - ' + str(list_unique[j])
                dict_stat[colname] = str(round(num, 3)) + ' (' + str(round(perc, 3)) + ') '
    return dict_stat
```

```python
stat_cols = ['age_enrollment', 'gender' , 'num_symptoms', 'num_comorbidities', 'median_sym_duration', 'infection_category', 'vax_covid_firstdose']
```

```python
dict_omicron = {}
df_omicron = df[df['VOC'] == 1]
print(df_omicron.shape[0])
dict_omicron = get_statistics(df_omicron, dict_omicron, stat_cols)
df_stats_omicron = pd.DataFrame.from_dict(dict_omicron, orient='index')
df_stats_omicron
```

```python
dict_pre_omicron = {}
df_pre_omicron = df[df['VOC'] == 0]
print(df_pre_omicron.shape[0])
dict_pre_omicron = get_statistics(df_pre_omicron, dict_pre_omicron, stat_cols)
df_stats_pre_omicron = pd.DataFrame.from_dict(dict_pre_omicron, orient='index')
df_stats_pre_omicron
```

```python
dict_total = {}
df_total = df.copy()
print(df_total.shape[0])
dict_total = get_statistics(df_total, dict_total, stat_cols)
df_stats_total= pd.DataFrame.from_dict(dict_total, orient='index')
df_stats_total
```

```python
# create a column with statistics (e.g. number of individuals in VOC=0, VOC=1), min max of age, etc.
# group by VOC and calculate the statistics
continues_cols = ["age_enrollment", "gender", 
                       'num_comorbidities', 'num_symptoms', 'median_sym_duration']

stats_VOC_0 = df[df["VOC"] == 0][continues_cols].describe()
stats_VOC_1 = df[df["VOC"] == 1][continues_cols].describe()

binary_columns = ['vax_flu_2019-2020', 'vax_flu_2020_2021', 'vax_covid_firstdose', 'hex_vax', 'at_least_one_comorbidity', 'infection_category']
percent_VOC_0 = df[df["VOC"] == 0][binary_columns].apply(lambda x: x.value_counts(normalize=True) * 100)
percent_VOC_1 = df[df["VOC"] == 1][binary_columns].apply(lambda x: x.value_counts(normalize=True) * 100)

stats_summary = pd.concat([stats_VOC_0, stats_VOC_1], axis=1, keys=["VOC_0", "VOC_1"])
percent_summary = pd.concat([percent_VOC_0, percent_VOC_1], axis=1, keys=["VOC_0", "VOC_1"])

# Mostra le statistiche
display(stats_summary) 
display(percent_summary)


```
